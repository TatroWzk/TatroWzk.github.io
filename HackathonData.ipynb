```python
import pandas as pd

file_path = '/content/Consulta-MARO-2024-10-25.xlsx'

excel_data = pd.ExcelFile(file_path)

sheet_names = excel_data.sheet_names
sheet_names
```

```python
df_empleo_servicios = excel_data.parse('Empleo en servicios')

df_empleo_servicios.head()
```

```python
df_empleo_servicios_cleaned = excel_data.parse('Empleo en servicios', skiprows=8)

df_empleo_servicios_cleaned.head()
```

```python
df_empleo_servicios_cleaned.columns = [
    "Subsector", "CIIU Rev.4", "Descripción CIIU Rev.4", "Departamento",
    "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022", "2023",
    "ENE-JUN 2023", "ENE-JUN 2024"
]

df_empleo_servicios_cleaned = df_empleo_servicios_cleaned.drop([0, 1, 2]).reset_index(drop=True)

df_empleo_servicios_cleaned.head()
```

```python
import pandas as pd
from prophet import Prophet

data_sucre = df_empleo_servicios_cleaned[df_empleo_servicios_cleaned['Departamento'] == 'Boyacá']

df_prophet = pd.melt(data_sucre,
                     id_vars=['Departamento'],
                     var_name='ds',
                     value_name='y')

df_prophet = df_prophet[['ds', 'y']].dropna()
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'], errors='coerce')
df_prophet['y'] = df_prophet['y'].replace(0, pd.NA)
df_prophet = df_prophet.dropna()

model = Prophet()
model.fit(df_prophet)

future = model.make_future_dataframe(periods=12, freq='M')
forecast = model.predict(future)

fig = model.plot(forecast)
```
